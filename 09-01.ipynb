## Concatenation

```python
import pandas as pd
```

```python
df1 =pd.read_csv('https://raw.githubusercontent.com/AlexTheAnalyst/PandasYouTubeSeries/refs/heads/main/LOTR.csv')
df1
```

```python
df2 =   pd.read_csv('https://raw.githubusercontent.com/AlexTheAnalyst/PandasYouTubeSeries/refs/heads/main/LOTR%202.csv')
df2
```

#concatenation 

```python
#pd.concat(objs, axis=0, join='outer', ignore_index=False, keys=None, ...)
```

```python
concat = pd.concat([df1,df2])
#concats in  axis=0 , by default,
#rows of df2 are added below df1
```

```python
concat
```

```python
concat =pd.concat([df1,df2],axis=0)
```

```python
concat
```

```python
concat =pd.concat([df1,df2],axis=0,ignore_index= True)
```

```python
concat
```

```python
concat = pd.concat([df1,df2],axis=1)
#columns form df2 are added to df1
```

```python
concat
```

```python
concat = pd.concat([df1,df2],axis=0,join='inner')
```

```python
concat
```
