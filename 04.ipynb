#SORTING in Pandas

```python
coffee_df.sort_values("Units Sold")
#sorts columns in increasing order  of units  sold values
```

```python
coffee_df.sort_values("Units Sold",ascending=False)
#sorts in decending order
```

```python
coffee_df.sort_values(["Units Sold","Coffee Type"])
#first  sorts by the unit sold and then coffe type
#if two or more rows have same value, then sorts them by their coffee type,inascending order.
```

```python
coffee_df.sort_values(["Units Sold","Coffee Type"],ascending= False)
```

```python
coffee_df.sort_values(["Units Sold","Coffee Type"],ascending= [0,1])
#here sorts the Units sold in decreasing order of values, if some rows have same value,
#sorts them  in ascending  alphabetical order.
#reminder : 0 ==> False
#         :1 ==>True
```
