#ADDING /REMOVING COLUMNS

```python
import pandas as pd
```

```python
coffee = pd.read_csv("./warmup-data/coffee.csv")
```

```python
coffee.head()
```

```python
coffee["price"] = 4.99 #adding a price column 
```

```python
coffee.head()
```

```python
import numpy as np
coffee["new_price"] =np.where(coffee["Coffee Type"]=="Espresso",3.99,5.99)
#we use np.where(x,y) to check if coffee type is espreeso, if its espresso prices is 3.99,
#if not 5.99
```

```python
coffee.head()
```

```python
coffee["Day"] = np.where(coffee["Day"]=="Saturday","OffDays",coffee["Day"])
#DataFrame["column_name"] = np.where(condition, value_if_true, value_if_false)

print(coffee )
#updates the day column in the data, if the coffee day is saturday, replaces it with offdays
```

```python
coffee["new_price"] = np.where(coffee["Day"]=="OffDays",1.99,coffee["new_price"])
#update the new_price to 1.99 for rows where the "Day" column is "OffDays"
print(coffee)
```

```python
coffee["price"] = np.where(coffee["Day"]=="OffDays",3.99,coffee["price"])
coffee
#updates the price column, if coffee day is offday, price is 3.99, or else same
```

```python
#if we want to drop the column

new_coffee_table=coffee.drop(columns=['price'])
print(new_coffee_table)
#this returns a new DataFrame with the column removed.,  it willl not remove the price 
#column unless explicitly  specified.
```
