```python
#filtering atheles with height greater than 215
bios.loc[bios['height_cm']>215]
```

```python
#filtering atheles with height greater than 215
bios.loc[bios['height_cm']>215, ["name","height_cm"]]
#bios.loc[row,column]
#row =>print all rows with height >215
#and for columns, display their name  and height_cm
```

```python
#for multiple conditions, we can add brackes
bios.loc[(bios['height_cm']>215) & (bios["born_country"]=="USA"), ["name","height_cm"]]
#row =>print all rows with height >215 & counttry  ==usa
#and for columns, display their name  and height_cm
```

```python
#filters based on strings operations
```

```python
bios[bios["name"].str.contains("Kim")]
#Kim is case senstive ,so must be capital K.
#filers athelete whose names containes kim
```

```python
bios[bios["name"].str.contains("kim",case =False)]
#using case =false, it accepts lower case.
```

```python
bios[bios["name"].str.contains("kim|jimmy",case =False)]
#filers athelete whose names containes kim or jimmy
```

```python
bios[bios["name"].str.contains("kim|jimmy",case =False ,regex=False)]
#wiill not return anything,because there is no exact match
```

```python
bios[bios["born_country"].isin(["USA","FRA"])]
#filters and provides athelete whos bonn country is in USA, France
```

```python
bios[bios["born_country"].isin(["USA","FRA"]) & (bios["name"].str.startswith("Jimmy"))]

#filters athettes whoes born country is either usa, or france, and name starts with Jimmy
```

```python
#filters athettes whoes born country is either usa, or germany, and name starts with Harry
bios[bios["born_country"].isin(["USA","GBR"]) & (bios["name"].str.startswith("Shaq"))]
```
